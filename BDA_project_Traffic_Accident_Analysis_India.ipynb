````python

# Setup: start SparkSession, imports, local paths
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, year, month, dayofmonth, hour, to_timestamp, when, unix_timestamp
from pyspark.sql.types import IntegerType, DoubleType
import pandas as pd, numpy as np, os, math
import matplotlib.pyplot as plt
# Matplotlib rule: one plot per figure, default colors (no explicit colors set).
plt.rcParams['figure.figsize'] = (10,5)

# Create directories for outputs
os.makedirs('data', exist_ok=True)
os.makedirs('figures', exist_ok=True)

# Path where notebook expects dataset (change if needed)
DATA_PATH = 'road_accidents.csv'  # <-- put your CSV here or change this path

# Create Spark session
spark = SparkSession.builder \
    .appName('TrafficAccidentAnalysis') \
    .config('spark.ui.showConsoleProgress','false') \
    .getOrCreate()

print('Spark session started.')
````

````python

# Robust CSV loader: tries a few common encodings and separator options
from pyspark.sql.utils import AnalysisException
def try_read_csv(path):
    options_list = [
        {'header':True, 'inferSchema':True, 'sep':','},
        {'header':True, 'inferSchema':True, 'sep':'\t'},
        {'header':True, 'inferSchema':True, 'sep':'|'},
    ]
    for opt in options_list:
        try:
            df = spark.read.option('header', str(opt['header']).lower()) \
                           .option('inferSchema', str(opt['inferSchema']).lower()) \
                           .option('sep', opt['sep']) \
                           .csv(path)
            # quick check: non-empty schema
            if len(df.columns) > 0:
                print('Loaded with sep:', opt['sep'])
                return df
        except Exception as e:
            # print('read attempt failed:', e)
            continue
    raise FileNotFoundError(f'Could not read CSV at {path} with common options.')

try:
    df = try_read_csv(DATA_PATH)
    print('Rows:', df.count(), 'Columns:', len(df.columns))
    display(df.limit(5).toPandas())
except FileNotFoundError as fe:
    print(fe)
    print('Notebook cannot continue without a dataset. Please place a CSV at the path or change DATA_PATH.')
    raise
````

````python

# Detect common column names (case-insensitive map)
cols = [c.lower().strip() for c in df.columns]
col_map = {c.lower():c for c in df.columns}  # maps lowercase->original name

def has_any(names):
    for n in names:
        if n.lower() in col_map:
            return col_map[n.lower()]
    return None

# Common columns
col_date = has_any(['date','accident_date','occurrence_date','crash_date'])
col_time = has_any(['time','accident_time','occurrence_time','crash_time','hour'])
col_state = has_any(['state','state/ut','state_ut','st_name'])
col_city  = has_any(['city','place','district','location','town'])
col_lat   = has_any(['latitude','lat'])
col_lon   = has_any(['longitude','lon','long','lng'])
col_fatal = has_any(['no_of_persons_killed','persons_killed','fatalities','no_of_fatalities','deaths','killed'])
col_inj   = has_any(['no_of_persons_injured','persons_injured','injuries','injured'])
col_sev   = has_any(['accident_severity','severity','type_of_accident','accident_type'])

print('Auto-detected columns:')
print('DATE:', col_date, 'TIME:', col_time, 'STATE:', col_state, 'CITY:', col_city)
print('LAT/LON:', col_lat, col_lon, 'FATAL:', col_fatal, 'INJURED:', col_inj, 'SEVERITY:', col_sev)

# Basic cleaning: drop exact duplicates, trim string columns
df = df.dropDuplicates()

string_cols = [c for c,t in df.dtypes if t=='string']
for c in string_cols:
    df = df.withColumn(c, when(col(c).isNotNull(), col(c)).otherwise(None))

# If date column exists try to parse it
if col_date:
    try:
        df = df.withColumn('accident_timestamp', to_timestamp(col(col_date)))
    except Exception as e:
        print('Could not parse date column to timestamp:', e)

# If time column exists and no parsed timestamp, try to create hour
if col_time and 'accident_timestamp' not in df.columns:
    # create hour if time-only column exists
    try:
        df = df.withColumn('hour', hour(to_timestamp(col(col_time), 'HH:mm')))
    except:
        pass

# If timestamp exists, extract year/month/day/hour
if 'accident_timestamp' in df.columns:
    df = df.withColumn('year', year(col('accident_timestamp')))
    df = df.withColumn('month', month(col('accident_timestamp')))
    df = df.withColumn('day', dayofmonth(col('accident_timestamp')))
    df = df.withColumn('hour', hour(col('accident_timestamp')))

print('After cleaning, columns count:', len(df.columns))
display(df.limit(5).toPandas())
````

````python

# Convert to pandas for plotting small aggregated tables
pandas_df = df.limit(1000000).toPandas() if df.count() > 0 else df.toPandas()  # limit if very large

# 1) Basic counts and distribution of years (if present)
if 'year' in df.columns:
    yearly = df.groupBy('year').count().orderBy('year').toPandas()
    plt.figure()
    plt.plot(yearly['year'], yearly['count'])
    plt.title('Yearly accident counts')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.grid(True)
    plt.savefig('figures/yearly_accidents.png')
    display(plt.gcf())

# 2) Top N states by accidents / fatalities
if '%s' in locals() and '%s' is not None:
    pass

# Detect state column name again for grouping
state_col = None
for c in df.columns:
    if c.lower() in ['state','state/ut','state_ut','st_name']:
        state_col = c
        break

if state_col:
    top_states = df.groupBy(state_col).count().orderBy(col('count').desc()).limit(15).toPandas()
    plt.figure()
    plt.bar(top_states[state_col], top_states['count'])
    plt.xticks(rotation=90)
    plt.title('Top states by number of accidents (top 15)')
    plt.ylabel('Accident count')
    plt.tight_layout()
    plt.savefig('figures/top_states_accidents.png')
    display(plt.gcf())

# 3) Fatalities distribution across states (if fatal column exists)
if '%s' in locals() and '%s' is not None:
    pass

fatal_col_name = None
for c in df.columns:
    if c.lower() in ['no_of_persons_killed','persons_killed','fatalities','no_of_fatalities','deaths','killed']:
        fatal_col_name = c
        break

if fatal_col_name and state_col:
    state_fatal = df.groupBy(state_col).sum(fatal_col_name).orderBy(col('sum({})'.format(fatal_col_name)).desc()).limit(15).toPandas()
    plt.figure()
    plt.bar(state_fatal[state_col], state_fatal['sum(%s)' % fatal_col_name])
    plt.xticks(rotation=90)
    plt.title('Top states by fatalities (top 15)')
    plt.tight_layout()
    plt.savefig('figures/top_states_fatalities.png')
    display(plt.gcf())

# 4) Hour-of-day heatmap (if hour exists)
if 'hour' in df.columns and state_col:
    # create pivot table: states x hour counts (limit to top 10 states)
    top_states_list = [r[0] for r in df.groupBy(state_col).count().orderBy(col('count').desc()).limit(10).select(state_col).collect()]
    pivot_df = df.where(col(state_col).isin(top_states_list)).groupBy(state_col, 'hour').count().toPandas()
    if not pivot_df.empty:
        pivot = pivot_df.pivot(index=state_col, columns='hour', values='count').fillna(0)
        plt.figure()
        plt.imshow(pivot.values, aspect='auto')
        plt.yticks(range(len(pivot.index)), pivot.index)
        plt.xlabel('Hour of day')
        plt.title('Accidents heatmap: top 10 states vs hour')
        plt.colorbar()
        plt.savefig('figures/state_hour_heatmap.png')
        display(plt.gcf())

# 5) Top causes / vehicle types (if such columns exist)
possible_cause_cols = [c for c in df.columns if 'cause' in c.lower() or 'reason' in c.lower()]
possible_vehicle_cols = [c for c in df.columns if 'vehicle' in c.lower() or 'affected_vehicle' in c.lower()]

if possible_cause_cols:
    ccol = possible_cause_cols[0]
    cause_counts = df.groupBy(ccol).count().orderBy(col('count').desc()).limit(15).toPandas()
    plt.figure()
    plt.bar(cause_counts[ccol].astype(str), cause_counts['count'])
    plt.xticks(rotation=90)
    plt.title('Top causes (sample)')
    plt.tight_layout()
    plt.savefig('figures/top_causes.png')
    display(plt.gcf())

if possible_vehicle_cols:
    vcol = possible_vehicle_cols[0]
    vehicle_counts = df.groupBy(vcol).count().orderBy(col('count').desc()).limit(15).toPandas()
    plt.figure()
    plt.bar(vehicle_counts[vcol].astype(str), vehicle_counts['count'])
    plt.xticks(rotation=90)
    plt.title('Top vehicle types involved (sample)')
    plt.tight_layout()
    plt.savefig('figures/top_vehicle_types.png')
    display(plt.gcf())

# 6) Correlation matrix (numerical columns)
num_cols = [c for c,t in df.dtypes if t in ('int','double','bigint','float')]
if len(num_cols) > 1:
    sample_pd = df.select(num_cols).dropna().limit(20000).toPandas()
    corr = sample_pd.corr()
    plt.figure()
    plt.imshow(corr.values, aspect='auto')
    plt.colorbar()
    plt.title('Correlation matrix (sample)')
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.tight_layout()
    plt.savefig('figures/correlation_matrix.png')
    display(plt.gcf())

print('EDA finished. Figures saved to /mnt/data/figures/ . You can include these in the final report.')
````

````python

# Optional ML: If a severity/fatal indicator exists, build simple classification models.
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator

# Find a candidate target column (binary or ordinal)
target_candidates = [c for c in df.columns if c.lower() in ['accident_severity','severity','fatal','fatality','is_fatal','death_flag']]
target = target_candidates[0] if target_candidates else None

if target:
    print('Using target column for modeling:', target)
    # If target is not numeric try to index it
    if dict(df.dtypes)[target] != 'int' and dict(df.dtypes)[target] != 'double':
        idx = StringIndexer(inputCol=target, outputCol='label').fit(df)
        df = idx.transform(df)
        label_col = 'label'
    else:
        df = df.withColumn('label', col(target).cast('double'))
        label_col = 'label'

    # select numerical features
    feature_cols = [c for c,t in df.dtypes if t in ('int','double','float','bigint') and c not in ('label','year','month','day','hour')]
    if not feature_cols:
        print('No numerical features found for ML. Skipping modeling.')
    else:
        assembler = VectorAssembler(inputCols=feature_cols, outputCol='features')
        data_ml = assembler.transform(df).select('features','label').na.drop()
        train, test = data_ml.randomSplit([0.8,0.2], seed=42)
        print('Train size:', train.count(), 'Test size:', test.count())

        # Logistic Regression
        lr = LogisticRegression(featuresCol='features', labelCol='label', maxIter=20)
        lr_model = lr.fit(train)
        preds_lr = lr_model.transform(test)
        evaluator = BinaryClassificationEvaluator()
        print('LR AUC:', evaluator.evaluate(preds_lr, {evaluator.metricName: 'areaUnderROC'}))

        # Random Forest
        rf = RandomForestClassifier(featuresCol='features', labelCol='label', numTrees=50)
        rf_model = rf.fit(train)
        preds_rf = rf_model.transform(test)
        print('RF AUC:', evaluator.evaluate(preds_rf, {evaluator.metricName: 'areaUnderROC'}))

        # Feature importance (from RF)
        try:
            importances = rf_model.featureImportances
            fi = list(zip(feature_cols, importances))
            fi_sorted = sorted(fi, key=lambda x: x[1], reverse=True)[:20]
            print('Top feature importances (RF):', fi_sorted)
        except Exception as e:
            print('Could not extract feature importances:', e)
else:
    print('No suitable target column found for modeling. Skipping ML section.')
````
